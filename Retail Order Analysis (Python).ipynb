```python
#making 'Not Available' and 'unknown' null
df = pd.read_csv('copy_of_orders.csv', na_values = ['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
df.head(15)
```

```python
df.columns
```

```python
#making column names all lowercase and join using underscore
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
#making discount column
df['discount'] = df['list_price']*df['discount_percent']/100
df.head(5)
```

```python
#making sell price column
df['sell_price'] =  df['list_price'] - df['discount']
df.head(5)
```

```python
#making profit column
df['profit'] =  df['sell_price'] - df['cost_price']
df.head(5)
```

```python
#all columns data types
df.dtypes
```

```python
#convert order_date from object type to datetime type
df['order_date'] = pd.to_datetime(df['order_date'], format = "%Y-%m-%d")
```

```python
df.dtypes
```

```python
df.drop(columns = ['cost_price','list_price','discount_percent'], inplace=True)
```

```python
df.drop(columns = ['unnamed:_0'], inplace=True)
```

```python
df.head(5)
```

```python
#replacing null values with unknown
df['ship_mode'] = df['ship_mode'].fillna('Unknown') 
df.head(15)
```
